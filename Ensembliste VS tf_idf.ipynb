```python
from evaluation_functions import plot_precision_recall, evaluate_all_datasets_tfidf, evaluate_all_datasets
import matplotlib.pyplot as plt
datasets = ['med']
```

# 3. Mod√®le ensembliste VS vectoriel

```python
results_tf = evaluate_all_datasets_tfidf(datasets, tf_idf=False)
results_tfidf = evaluate_all_datasets_tfidf(datasets, tf_idf=True)
results_ensembliste = evaluate_all_datasets(datasets)
```

```python
results_linear = dict()
results_linear["ensembliste"] = results_ensembliste["med"]["evaluation_linear"]
results_linear["tf"] = results_tf["med"]["evaluation_linear"]
results_linear["tf_idf"] = results_tfidf["med"]["evaluation_linear"]

results_inverted = dict()
results_inverted["ensembliste"] = results_ensembliste["med"]["evaluation_inverted"]
results_inverted["tf"] = results_tf["med"]["evaluation_inverted"]
results_inverted["tf_idf"] = results_tfidf["med"]["evaluation_inverted"]
```

```python
plt.figure(figsize=(6,10))
plt.subplot(211)
plot_precision_recall([results_linear[e] for e in results_linear], [e for e in results_linear])
plt.title("med dataset SRI comparison with linear index")
plt.subplot(212)
plot_precision_recall([results_inverted[e] for e in results_linear], [e for e in results_inverted])
plt.title("med dataset SRI comparison with inverted index")
plt.savefig("ensembliste_vs_tfidf.png")
plt.show()
```
