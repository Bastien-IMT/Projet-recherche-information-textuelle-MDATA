```python
from evaluation_functions import evaluate_all_datasets, plot_performances, merge_results, plot_performances_comparison, compare_time

datasets = ['cacm', 'cisi', 'cran', 'med', 'lisa', 'time']
# datasets = ['cacm', 'cisi', 'cran', 'med']
subplot = 320
```

# 1. Mod√®le ensembliste
## 1.1. Data fully preprocessed
(Without words in stop list and SnowballStemmer applied)

```python
results_fully_preprocessed = evaluate_all_datasets(datasets)
plot_performances(results_fully_preprocessed, subplot_value=subplot, title="Recall-precision curves with ensembliste SRI data fully preprocessed.", save_plot=True)
```

## 1.2. Data with stop words and with stemmer

```python
results_stopwords = evaluate_all_datasets(datasets, stop_words=False)
origin_stopwords_results = merge_results(results_fully_preprocessed, results_stopwords)
plot_performances_comparison(origin_stopwords_results, subplot_value=subplot, title="Recall-precision curves comparison with ensembliste SRI data fully preprocessed VS data with stopwords.", save_plot=True)
```

## 1.3. Data without stop words and without stemmer

```python
results_stemmer = evaluate_all_datasets(datasets, stemm=False)
origin_stemmer_results = merge_results(results_fully_preprocessed, results_stemmer)
plot_performances_comparison(origin_stemmer_results, subplot_value=subplot, title="Recall-precision curves comparison with ensembliste SRI data fully preprocessed VS data without stemmer.", save_plot=True)
```

## 1.4. Data with stop words and without stemmer

```python
results_raw_data = evaluate_all_datasets(datasets, stop_words=False, stemm=False)
origin_raw_results = merge_results(results_fully_preprocessed, results_raw_data)
plot_performances_comparison(origin_stemmer_results, subplot_value=subplot, title="Recall-precision curves comparison with ensembliste SRI data fully preprocessed VS raw data.", save_plot=True)
```

```python
all_results = {"results_fully_preprocessed": results_fully_preprocessed,
               "results_stopwords": results_stopwords,
               "results_stemmer" :results_stemmer,
               "results_raw_data" :results_raw_data}


time_comp = compare_time(datasets, all_results)

```
